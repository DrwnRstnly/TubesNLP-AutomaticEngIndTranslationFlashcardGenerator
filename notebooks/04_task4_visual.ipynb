```python
import numpy as np
def plot_qwen_metrics(data):
    labels = [row[0] for row in data]
    bleu = [row[1] for row in data]
    rouge = [row[2] for row in data]
    bert = [row[3] for row in data]

    x = np.arange(len(labels))
    width = 0.25

    plt.figure(figsize=(20, 7))
    
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.bar(x - width, bleu, width, label="BLEU")
    plt.bar(x, rouge, width, label="RougeL")
    plt.bar(x + width, bert, width, label="BERT")

    plt.xticks(x, labels, rotation=75, ha="right")
    plt.ylabel("Score")
    plt.title("Metric Comparison for All Qwen Models")
    plt.legend()
    plt.tight_layout()
    return plt
```

```python
filtered_df = df_main[ df_main["similarity_score"] > 0.6 ].reset_index(drop=True)
```

```python

# Hitung jumlah kata
filtered_df["word_count"] = filtered_df["english_sentence"].apply(lambda x: len(str(x).split()))
filtered_df["word_count_2"] = filtered_df["indonesian_sentence"].apply(lambda x: len(str(x).split()))

# Plot histogram
plt.figure(figsize=(10, 6))

plt.hist(filtered_df["word_count"], bins=200, alpha=0.6, label="English", color="blue")
plt.hist(filtered_df["word_count_2"], bins=200, alpha=0.6, label="Indonesian", color="orange")

plt.xlabel("Number of Words per Sentence")
plt.ylabel("Frequency")
plt.title("Distribution of Word Counts (English vs Indonesian)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
df_short = filtered_df[(filtered_df["word_count"] < 200) & 
                       (filtered_df["word_count_2"] < 200)]

plt.figure(figsize=(10,6))

plt.hist(df_short["word_count"], bins=100, alpha=0.6, label="English", color="blue")
plt.hist(df_short["word_count_2"], bins=100, alpha=0.6, label="Indonesian", color="orange")

plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Distribution of Word Counts (<200 words only)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
