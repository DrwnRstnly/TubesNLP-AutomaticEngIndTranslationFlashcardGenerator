```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../src/task1_ranker/results/eval_finetuned_and_base.csv")
df
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="model", y="pearson", hue="type")
plt.xticks(rotation=45)
plt.ylim(0.80, 0.88) 
plt.title("Pearson Correlation Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="model", y="spearman", hue="type")
plt.xticks(rotation=45)
plt.ylim(0.80, 0.88) 
plt.title("Spearman Correlation Comparison")
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.scatterplot(
    data=df,
    x="pearson",
    y="spearman",
    hue="type",
    palette={"base": "blue", "full_finetuned": "green", "lora_finetuned": "red"},
    s=140
)

for i, row in df.iterrows():
    plt.text(row["pearson"] + 0.0003, row["spearman"] + 0.0003, row["model"], fontsize=9)

plt.title("Pearson vs Spearman  Annotated by Model")
plt.xlim(0.82, 0.88)
plt.ylim(0.82, 0.88)
plt.show()
```
